```python
import pandas as pd
import requests
import json
from config import gkey
from pprint import pprint
import numpy as np
```

```python
#Grab all the beer review data from file
beer_review_data=pd.read_csv('Resources/beer_reviews.csv', encoding="utf-8")
beer_review_data
```

```python
#Find the average value for the reviews by the beer name
beer_group=beer_review_data.groupby(["beer_beerid","beer_name","brewery_name","beer_style","brewery_id"])
beer_df=pd.DataFrame(beer_group["review_overall","review_aroma","review_appearance","review_taste","review_palate","beer_abv"].mean())

beer_df=beer_df.reset_index()

beer_df
```

```python
#Get a count of the number of reviews for each beer
review_count=beer_review_data.groupby(["beer_beerid"]).count()
review_count=review_count[["review_overall"]]
review_count.head()
```

```python
#Merge average review dataframe and review count dataframe
beer_merged_data = pd.merge(beer_df, review_count, on='beer_beerid', how='outer')
beer_merged_data = beer_merged_data.rename(columns={'review_overall_x': 'review_overall', 'review_overall_y': 'review_count'})


beer_merged_data.head()
```

```python
#filter out those with less than 50 reviews
beer_50review=beer_merged_data.loc[(beer_merged_data["review_count"]>50),:]
beer_50review
```

```python
#Write file with all beer with above 50 reviews
beer_50review.to_csv('Output/beer_review_avg_worldwide_try2.csv')
```

```python
#Create new data frame for breweries
brewery_df= beer_50review.groupby(["brewery_name", "brewery_id"])
#find brewery rating based on average beer review
average_rating = round(brewery_df['review_overall'].mean(), 2)
#get the total count of reviews made on beers from each brewery
review_count=brewery_df["review_count"].sum()

brewery_df = pd.DataFrame(average_rating)
brewery_df["review_count"]=review_count
brewery_df = brewery_df.reset_index()
brewery_df = brewery_df.rename(columns={'brewery_name': 'Brewery'})

brewery_df

#brew_test=brewery_df.head(10)
```

```python
#create new columns to save stuff from api
brewery_df["lat"]=''
brewery_df["lng"]=''
brewery_df["address"]=''
```

<h1>Time for some google api shenanigans</h1>

```python
#setup fields that I want from google and basic
field="geometry/location,formatted_address"
base_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"

#set up for loop to go through every row of my brewery dataframe
for index, row in brewery_df.iterrows():
    params = {
        "input": row['Brewery'],
        "key": gkey,
        "inputtype":"textquery",
        "fields": field
    }
    
    top_response = requests.get(base_url, params=params)
    result = top_response.json()
    try:
        #attempt to save data from google
        brewery_df.loc[index, 'lat'] = result['candidates'][0]['geometry']['location']['lat']
        brewery_df.loc[index, 'lng'] = result['candidates'][0]['geometry']['location']['lng']
        brewery_df.loc[index, 'address'] = result['candidates'][0]['formatted_address']
    
    except (KeyError, IndexError):
        #output lack of results from query
        print(f"No Google result for {row['Brewery']}......Skipping.....")
```

```python
brewery_df
```

```python
#fill empty locations with Nan and then drop
brewery_df=brewery_df.replace(r'^\s*$', np.NaN, regex=True)
brewery_df.dropna(inplace = True)
brewery_df
```

```python
#create new coulmns to save a split of the address column
brewery_df["street_address"]=""
brewery_df["city"]=""
brewery_df["postal_code"]=''
brewery_df["country"]=''
brewery_df.head()
```

```python
for index, row in brewery_df.iterrows():
    #split and reverse so country is first
    addresssplit=row["address"].split(", ")
    addresssplit.reverse()
    try:
        #attempt to save values to each location
        brewery_df.loc[index, 'country'] = addresssplit[0]
        brewery_df.loc[index, 'postal_code'] = addresssplit[1]
        brewery_df.loc[index, 'city'] = addresssplit[2]
        brewery_df.loc[index, 'street_address']=addresssplit[3]
    except(IndexError):
        #output errors
        print(f"funky address for {row['Brewery']}")
brewery_df.head(10)
```
