```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
beer_df = pd.read_csv("Resources/clean_beer_data_try2.csv")
beer_df.head()
```

```python
plt.scatter(beer_df["review_aroma"],beer_df["review_overall"])
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(beer_df["review_aroma"], beer_df["review_overall"])
regress_values = beer_df["review_aroma"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(beer_df["review_aroma"],regress_values,"r-")
plt.annotate(line_eq,(3.5,2.0),fontsize=15,color="red")
plt.xlabel("Aroma out of 5")
plt.ylabel("Review Overall out of 5")
plt.title("Overall Vs Aroma")
print(f"{round(rvalue,2)}")
```
